### **2da Pregunta:** ¿Cuál fue la facturación por categoria em 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Calcular la facturación por línea de compra
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']

# Agrupar por categoría y sumar la facturación total
facturacion_por_categoria = df.groupby('categoria')['facturacion'].sum().reset_index()

# Ordenar de mayor a menor
facturacion_por_categoria = facturacion_por_categoria.sort_values(by='facturacion', ascending=True)

# Crear el gráfico de barras horizontales
plt.figure(figsize=(10, 6))
sns.barplot(data=facturacion_por_categoria, x='facturacion', y='categoria', palette='Blues_d')

plt.title('Facturación Total por Categoría')
plt.xlabel('Facturación MXN$')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentarios: </font>

---

Aquí especificamos el tipo de gráfico, guiando a ChatGPT en la creación de nuestro visual. Podemos notar que en un gráfico de barras el enfoque está más centrado en las categorías, además de ser una forma más fluida de leer textos que podrían ser demasiado extensos para un gráfico de columnas.


### **3a Pregunta:** ¿Cómo fueron las ventas a lo largo de los meses en 2023?


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de líneas que muestre las ventas totales mensuales de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturación: calculada como el producto del precio unitario y la cantidad, sumado al costo de envío de cada venta.

Deseamos crear una visualización que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del índice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gráfico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}
```


Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Crear una columna 'mes' con el nombre del mes en inglés
df['mes'] = df['fecha'].dt.month_name()

# Agrupar por mes (utilizando el primer día del mes como referencia)
ventas_mensuales = df.groupby(df['fecha'].dt.to_period('M'))['facturacion'].sum().reset_index()
ventas_mensuales['fecha'] = ventas_mensuales['fecha'].dt.to_timestamp()

# Diccionario de traducción de los meses
meses = {
    'January': 'Ene', 'February': 'Feb', 'March': 'Mar', 'April': 'Abr',
    'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
    'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
}

# Traducir los nombres de los meses
ventas_mensuales['mes'] = ventas_mensuales['fecha'].dt.month_name().map(meses)

# Crear gráfico de líneas
plt.figure(figsize=(10, 5))
sns.lineplot(data=ventas_mensuales, x='mes', y='facturacion', marker='o', sort=False)

plt.title('Ventas Totales Mensuales de Zoop')
plt.xlabel('Mes')
plt.ylabel('Facturación Total')
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentarios: </font>

---

Cuando tenemos una cantidad mayor de períodos en nuestros datos, vale la pena recurrir a gráficos de líneas, que son ideales para series temporales. Dado que nuestros datos son mensuales, utilizar un gráfico de columnas, por ejemplo, reduciría la percepción de la secuencia de los datos. Las líneas ofrecen una mayor sensación de continuidad.

##**2.3 - Componiendo datos dinámicos**


Construimos algunos visuales interesantes para comparar los datos, mostrando las facturaciones de Zoop. Ahora, pasemos a una nueva categoría de visuales: composición.

Vamos a hacer una composición dinámica de los datos. Para ello, añadiremos más variables al análisis, incorporando 3 variables diferentes, siendo una de ellas categórica.

###**4a Pregunta:** ¿Cómo está dividida la venta por trimestre por método de pago? ¿Existe algún período en el que algún tipo de pago se destaque?


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras apiladas que muestre las ventas por trimestre en relación con el método de pago de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python


# Crear una columna con el trimestre en formato "Q1-2023", "Q2-2023", etc.
df['trimestre'] = df['fecha'].dt.to_period('Q').astype(str)

# Agrupar por trimestre y método de pago, sumando la facturación
ventas_trimestrales = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().unstack()

ventas_trimestrales
````

````python

# Crear gráfico de barras apiladas
ventas_trimestrales.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='Set2'
)

plt.title('Método de Pago por Trimestre de 2023')
plt.xlabel('Trimestre')
plt.ylabel('Facturación')
plt.xticks(rotation = 45)
plt.legend(title='Método de Pago', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentarios: </font>

---

El gráfico de columnas apiladas añade a la comparación entre períodos o categorías, la comparación también entre los datos de cada período/categoría. Es como si pusiéramos una "lupa" en cada columna y visualizáramos la composición de cada una de acuerdo con la variable deseada (aquí, las formas de pago).

## **4.4 - Agregar elementos visuales**

¿Qué tal el uso de textos informativos, verdad? ¿Y si ahora añadimos otros elementos visuales a uno de nuestros gráficos?

Vamos a elegir ahora el gráfico de líneas (Ventas totales mensuales de Zoop) y aplicar una técnica más a él mediante el uso de la IA.


````python
# Crear una columna 'mes' con el nombre del mes en inglés
df['mes'] = df['fecha'].dt.month_name()

# Agrupar por mes (utilizando el primer día del mes como referencia)
ventas_mensuales = df.groupby(df['fecha'].dt.to_period('M'))['facturacion'].sum().reset_index()
ventas_mensuales['fecha'] = ventas_mensuales['fecha'].dt.to_timestamp()

# Diccionario de traducción de los meses
meses = {
    'January': 'Ene', 'February': 'Feb', 'March': 'Mar', 'April': 'Abr',
    'May': 'May', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
    'September': 'Sep', 'October': 'Oct', 'November': 'Nov', 'December': 'Dic'
}

# Traducir los nombres de los meses
ventas_mensuales['mes'] = ventas_mensuales['fecha'].dt.month_name().map(meses)

# Crear gráfico de líneas
plt.figure(figsize=(10, 5))
sns.lineplot(data=ventas_mensuales, x='mes', y='facturacion', marker='o', sort=False)

plt.title('Ventas Totales Mensuales de Zoop')
plt.xlabel('Mes')
plt.ylabel('Facturación Total')
plt.tight_layout()
plt.show()
````

```python
f"Promedio de calificaciones = {df['nota'].mean().round(2)}\n Calificación más común = {df['nota'].value_counts().idxmax()}\n Frecuencia más común = {df['nota'].value_counts().max()}"
```

en color AZUL_1. Añade la caja de anotación con el texto en la posición xytext=(2, 1800) y coloca una flecha, en color AZUL_1, saliendo de la caja hasta la posición xy=(8.7, 2250).

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
def distribucion_calificaciones():
  # Colores de identidad visual de Zoop
  GRIS_1 = "#f2f2f2"
  GRIS_3 = "#cccccc"
  GRIS_5 = "#4d4d4d"
  AZUL_1 = "#1890FF"
  ROJO_1 = "#FF4D4F"
  BLANCO = "#ffffff"

  # Crear figura y ejes
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Histograma
  sns.histplot(df['nota'], bins=11, kde=False, color=ROJO_1, ax=ax)

  # Título
  ax.set_title("Distribución de las evaluaciones de las compras en 2023", fontsize=18, color=GRIS_5)

  # Ejes
  ax.set_xlabel("Calificación (0-10)", fontsize=12, color=AZUL_1)
  ax.set_ylabel("Frecuencia", fontsize=12, color=AZUL_1)
  ax.set_xticks(range(11))
  ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

  # Borrar bordes
  for spine in ['top', 'bottom', 'left', 'right']:
      ax.spines[spine].set_visible(False)

  # Grid
  ax.set_axisbelow(True)
  ax.yaxis.grid(True, linestyle='--', color=GRIS_3)

  # Caja de anotación
  promedio = df['nota'].mean().round(2)
  nota_moda = df['nota'].value_counts().idxmax()
  frecuencia_moda = df['nota'].value_counts().max()
  texto = (
      f"Promedio de calificaciones = {promedio}\n"
      f"Calificación más común = {nota_moda}\n"
      f"Frecuencia más común = {frecuencia_moda}"
  )

  ax.annotate(
      texto,
      xy=(8.7, 2250),
      xytext=(2, 1800),
      textcoords='data',
      fontsize=12,
      color=AZUL_1,
      bbox=dict(boxstyle="round,pad=0.5", facecolor=BLANCO, edgecolor=GRIS_3),
      arrowprops=dict(arrowstyle="->", color=AZUL_1)
  )

  plt.tight_layout()
  return plt
````

````python
def box_plot_edad_sexo():
  # Colores Zoop
  GRIS_1 = "#f2f2f2"
  GRIS_3 = "#cccccc"
  GRIS_5 = "#4d4d4d"
  AZUL_1 = "#1890FF"
  AQUA_1 = "#40E0D0"
  BLANCO = "#ffffff"

  # Crear figura y ejes
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Boxplot
  sns.boxplot(x='sexo_biologico', y='edad', data=df, ax=ax, palette=[AZUL_1, AQUA_1])

  # Título
  ax.set_title("Distribución de la Edad por Sexo Biológico de los clientes", fontsize=18, color=GRIS_5)

  # Ejes
  ax.set_xlabel("")
  ax.set_ylabel("Edad", fontsize=12, color=AZUL_1)
  ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

  # Quitar bordes
  for spine in ['top', 'bottom', 'left', 'right']:
      ax.spines[spine].set_visible(False)

  # Cuadrícula
  ax.set_axisbelow(True)
  ax.yaxis.grid(True, linestyle='--', color=GRIS_3)

  # Anotaciones
  # Para sexo_biologico = 1 (asumimos índice 1 es "Masculino", index 0 es "Femenino" o viceversa)
  grupo = df.groupby("sexo_biologico")["edad"]
  valores1 = grupo.min()[1], grupo.quantile(0.25)[1], grupo.median()[1], grupo.quantile(0.75)[1], grupo.max()[1]
  valores0 = grupo.min()[0], grupo.quantile(0.25)[0], grupo.median()[0], grupo.quantile(0.75)[0], grupo.max()[0]

  texto1 = (
      f"Mínimo = {valores1[0]}\n"
      f"1º quartil = {valores1[1]}\n"
      f"2º quartil = {valores1[2]}\n"
      f"3º quartil = {valores1[3]}\n"
      f"Máximo = {valores1[4]}"
  )

  texto0 = (
      f"Mínimo = {valores0[0]}\n"
      f"1º quartil = {valores0[1]}\n"
      f"2º quartil = {valores0[2]}\n"
      f"3º quartil = {valores0[3]}\n"
      f"Máximo = {valores0[4]}"
  )

  # Anotación sexo índice 1
  ax.annotate(
      texto1,
      xy=(0.9, 45),
      xytext=(-0.4, 50),
      textcoords='data',
      fontsize=12,
      color=AZUL_1,
      bbox=dict(boxstyle="round,pad=0.5", facecolor=BLANCO, edgecolor=GRIS_3),
      arrowprops=dict(arrowstyle="->", color=AZUL_1)
  )

  # Anotación sexo índice 0
  ax.annotate(
      texto0,
      xy=(0.1, 45),
      xytext=(0.6, 50),
      textcoords='data',
      fontsize=12,
      color=AZUL_1,
      bbox=dict(boxstyle="round,pad=0.5", facecolor=BLANCO, edgecolor=GRIS_3),
      arrowprops=dict(arrowstyle="->", color=AZUL_1)
  )

  plt.tight_layout()
  return plt
````
